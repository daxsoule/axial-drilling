# grdimage turns all black

With one region extents grdimage works fine:

```python
import pygmt
fig = pygmt.Figure()
fig.basemap(region=[-130-0.3/60, -130+2/60, 45+54/60, 45+55/60], projection="M-130/46/7.5i", frame=True)
fig.grdimage('../data/axial_auv_mbari09and11s_gcs_1m_v3.grd', V='d')
fig.show()
```
